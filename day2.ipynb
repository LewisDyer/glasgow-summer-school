### Visualising data using matplotlib

In this notebook you are going to see how to generate some of the plots we've seen in the lecture. 

The documentation of matplotlib is really good, so worth having a look if you get stuck.

This notebook covers the following examples:

* Example 1: Quarterly gas measurements, UK 1960-1986

* Example 1: Improving the earlier plot and splitting it by season

* Example 2: Visualising disributions 

* Example 3: Bar chart

```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
plt.rc('figure', figsize=(8.0, 4.0), dpi=180)
```

```python
gas = np.loadtxt("data/gas.csv", delimiter=',')
```

```python
plt.plot(gas[:, 1], gas[:, 2], "o-") #plt.scatter is another option here but we can't draw a line graph with it
plt.ylabel("Gas consumed (megatherms)")
plt.grid("on") #sometimes keeping the grid on makes the oplot more readable 
plt.ylim(0, 1500)
plt.title("UK Gas quarterly gas usage 1960-1986")
```

```python
# split into quarterly lines
q1 = gas[::4, :]
q2 = gas[1::4, :]
q3 = gas[2::4, :]
q4 = gas[3::4, :]

#markers and colours are useful here to help us distinguish between the different seasons

plt.plot(q1[:,1], q1[:,2], label='Winter', marker='d') 
plt.plot(q1[:,1], q2[:,2], label='Spring', marker='o')
plt.plot(q1[:,1], q3[:,2], label='Summer', marker='*')
plt.plot(q1[:,1], q4[:,2], label='Autumn', marker='^')

plt.xlabel('Date (years AD)')
plt.ylabel('Gas consumed (megatherms)')
plt.grid("on")
plt.legend()
plt.ylim(0,1500)
plt.title("UK Gas quarterly gas usage 1960-1986")
```

### Histograms

* Shows count of values within a range. 
* Used to visualise distributions
* Number of bins is important 

```python
x = gas[:,2]
plt.hist(x, bins=np.linspace(0,1200,20));
plt.xlabel("Megatherms")
plt.ylabel("Count of quarters")
```

### Bar charts

While they might look a bit like histograms, there are some main differences: 

* In histograms each column represents a group defined by a continuous variable 
* In bar charts, each column is a group defined by a categorical variable
* A histogram describes a distribution
* Order of categorical variables on x axis doesn't matter in a bar chart 

```python
dino = ['T-rex','V-raptor','Brachiosaurus']
y_pos = np.arange(len(dino))
arms = [4,7,12]
 
plt.bar(y_pos,arms,color='green')

plt.xticks(y_pos, dino)
plt.ylabel("Arms length")
plt.title("Length of Dinosaur Arms")
plt.show()
```
